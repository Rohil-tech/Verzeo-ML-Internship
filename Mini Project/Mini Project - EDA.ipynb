```python
import pandas as pd
```

```python
movies_df = pd.read_csv('tmdb-movies.csv')
```

```python
movies_df
```

```python
movies_df.columns
```

```python
movies_df = movies_df[[ 'original_title', 'budget', 'revenue', 'cast',
        'genres', 'release_year', 'runtime']]
```

```python
movies_df
```

```python
movies_df = movies_df.dropna()
```

```python
movies_df = movies_df.drop_duplicates()
```

```python
movies_df
```

```python
movies_df = movies_df[(movies_df['budget'] != 0) & (movies_df['revenue'] != 0)
                     & (movies_df['runtime'] != 0)]
```

```python
movies_df
```

```python
df_title = movies_df[(movies_df['release_year'] >= 2000) & 
                     (movies_df['release_year'] <= 2005)]
```

```python
df_title
```

```python
number_of_words = {}
for i in range(df_title.shape[0]):
    word = str(df_title['original_title'].values[i]).split(' ')
    for j in word:
        try:
            count = number_of_words[j]
            number_of_words[j] = count + 1
        except:
            number_of_words[j] = 1
            
number_of_words
```

```python
total = 0
for value in number_of_words.values():
    total += value
    
total
```

```python
rows = df_title.shape[0]
```

```python
total/rows
```
